```python
def is_primitive_root(u, q, factors):
    phi = q - 1
    for p in factors:
        # check u^(phi/p) mod q != 1
        if pow(u, phi // p, q) == 1:
            return False
    return True

def find_primitive_root(q, factors):
    for u in range(2, q):
        if is_primitive_root(u, q, factors):
            return u
    return None
```

```python
q = 577
factors = [2, 3]
primitive_root = find_primitive_root(q, factors)
print("Primitive root modulo", q, "is:", primitive_root)
```

```python
def calculate_omega(r, k, q):
    return pow(r, k, q)

print(calculate_omega((99 - (pow(136, 0) * 3) // 2), 1, 137))
```

```python
def mod_inverse(a, n):
    return pow(a, -1, n)

print(mod_inverse(171, 577))
```
